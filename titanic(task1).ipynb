```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

ship = pd.read_csv('https://github.com/Prabodh-09/CODSOFT/raw/main/Titanic-Dataset%20(1).csv')
ship.head()


```

```python
#viewing data
```

```python
ship.info()
ship.shape
ship.describe()
```

```python
ship['Survived'].value_counts()
```

```python
sns.countplot(x='Survived',data=ship)
```

```python
sns.countplot(x=ship['Survived'],hue=ship['Sex'])
```

```python
ship['Sex'] = ship['Sex'].replace({'male': 0, 'female': 1})
ship['Embarked'] = ship['Embarked'].replace({'S': 0, 'C': 1, 'Q': 2})
ship.head()
```

```python
#preprocessing data
```

```python
ship['Age'].fillna(ship['Age'].median(), inplace=True)
ship['Embarked'].fillna(ship['Embarked'].mode()[0], inplace=True)
ship.dropna(axis=1, inplace=True)
ship.head()
```

```python
#now splitting variales into dependent and independent
```

```python
x=ship[['Age','Sex','Embarked','Pclass']]
y=ship['Survived']
```

```python
plt.figure(figsize=(10,6))
plt.bar(ship['Pclass'],ship['Survived'])
plt.xlabel('Pclass')
plt.ylabel('Survived')
plt.title('Pclass vs Survived')
plt.show()
```

```python
#splitting into training and testing
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
x_train
```

```python
#Model selection
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
#training
```

```python
model.fit(x_train,y_train)
```

```python
#testing
```

```python
print(model.predict(x_test))
```

```python
y_test[0:10]
```

```python
ypred=model.predict(x_test)
```

```python
#accuracy
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,ypred)
```

```python
#accuracy sore=79%
```

```python
def get_passenger_info(passenger_id):
    passenger = ship[ship['PassengerId'] == passenger_id]
    if not passenger.empty:
        name = passenger.iloc[0]['Name']
        survived = passenger.iloc[0]['Survived']
        survived_status = "Survived" if survived == 1 else "Did not survive"
        return f"Passenger Name: {name}, Survival Status: {survived_status}"
    else:
        return "Passenger ID not found."


passenger_id = 20
print(get_passenger_info(passenger_id))
```
